```python
%matplotlib inline
```

```python
# Import dependencies
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollar accumulated for each month
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create the plot
plt.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
ax = plt.axes()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot
plt.plot(x_axis, y_axis)
plt.show()
```

```python
# Create the plot and add a label for the legend
plt.plot(x_axis, y_axis, label="Boston")

# Create labels for the x and y axes
plt.xlabel("Date")
plt.ylabel("Fare($)")

# Set the y limit between 0 and 45
plt.ylim(0, 45)

# Create a title
plt.title("PyBer Fare by Month")

# Add the legend
plt.legend()
```

```python
# Create the plot
plt.plot(x_axis, y_axis, marker="*", color="blue", linewidth=2, label="Boston")

# Create labels for the x and y axes
plt.xlabel("Date")
plt.ylabel("Fare($)")

# Set the y limit between 0 and 45
plt.ylim(0, 45)

# Create a title
plt.title("PyBer Fare by Month")

# Add a grid
plt.grid()

# Add the legend
plt.legend()
```

```python
# Formatting using object-oriented interface
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis, marker="*", color="green", linewidth=2, label="Boston")
ax.set_title("PyBer Fare by Month")
ax.set_xlabel("Date")
ax.set_ylabel("Fare($)")
ax.set_ylim(0, 45)
ax.grid()
```

```python
# Create the plot
plt.bar(x_axis, y_axis)
```

```python
# Create the plot
plt.bar(x_axis, y_axis, color="green", label="Boston")

# Create labels for the x and y axes
plt.xlabel("Date")
plt.ylabel("Fare($)")

# Create a title
plt.title("PyBer Fare by Month")

# Set the y axis limit
plt.ylim(0, 45)

# Add the legend
plt.legend()
```

```python
# Create the plot
plt.barh(x_axis, y_axis, color="magenta", label="Boston")
plt.gca().invert_yaxis()
plt.title("PyBer Fare by Month")
plt.xlabel("Fare($)")
plt.ylabel("Month")
```

```python
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.bar(x_axis, y_axis, color="k", label="New York")
```

```python
# Create the plot with ax.plot()
fig, ax = plt.subplots()
ax.barh(x_axis, y_axis, color="cyan", label="Chicago")
ax.set_title("PyBer Fare by Month")
ax.set_ylabel("Month")
ax.set_xlabel("Fare($)")
ax.invert_yaxis()
```

```python
gix, ax = plt.subplots()
ax.scatter(y_axis, x_axis, linewidth=2, edgecolor="black", alpha=0.8, label="Boston", c="skyblue", s=[i*5 for i in y_axis])
ax.legend()
ax.set_xlabel("Fare($)")
ax.set_ylabel("Date")
ax.set_title("PyBer Fare by Month")
ax.invert_yaxis()
ax.set_xlim(0,50)
```

```python
plt.pie(y_axis, labels=x_axis)
plt.show()
```

```python
plt.subplots(figsize=(8, 8))
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.pie(y_axis, explode=explode_values, labels=x_axis, autopct="%.1f%%")
plt.show()
```

```python
# Assign 12 colors, one for each month
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.subplots(figsize=(8, 8))
plt.pie(y_axis, explode=explode_values, colors=colors, labels=x_axis, autopct="%.1f%%")
plt.show()
```

```python
exploding_values = (0, 0, 0.2, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
fig, ax = plt.subplots(figsize=(8, 8))
ax.pie(y_axis, colors=colors, counterclock=False, startangle=90, labels=x_axis, shadow=True, autopct="%.1f%%", explode=exploding_values)
plt.show()
```
